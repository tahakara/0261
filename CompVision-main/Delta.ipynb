### Thresholding

```python
a= cv2.imread('DATA/2_rgb.png')
a= cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
plt.imshow(a)
```

```python
orijinalResim = cv2.imread('DATA/2_rgb.png',0)
# orijinalResim = cv2.cvtColor(orijinalResim, cv2.COLOR_BGR2RGB)

ret1, sonuc1 = cv2.threshold(orijinalResim,127,255,cv2.THRESH_BINARY)
ret2, sonuc2 = cv2.threshold(orijinalResim,127,0,cv2.THRESH_TRUNC)
ret3, sonuc3 = cv2.threshold(orijinalResim,127,0,cv2.THRESH_TOZERO)

f, eksen = plt.subplots(4,1)
eksen[0].imshow(orijinalResim, cmap="gray")
eksen[1].imshow(sonuc1, cmap="gray")
eksen[2].imshow(sonuc2, cmap="gray")
eksen[3].imshow(sonuc3, cmap="gray")
```

```python
orijinalResim = cv2.imread('DATA/5_gazete.png',0)
sonuc1 = cv2.adaptiveThreshold(orijinalResim,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,55,1)
sonuc2 = cv2.adaptiveThreshold(orijinalResim,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,55,1)
sonuc3 = cv2.adaptiveThreshold(orijinalResim,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,55,1)

f, eksen = plt.subplots(2,2)
eksen[0,0].imshow(orijinalResim, cmap="gray")
eksen[0,1].imshow(sonuc1, cmap="gray")
eksen[1,0].imshow(sonuc2, cmap="gray")
eksen[1,1].imshow(sonuc3, cmap="gray")
```

```python
resim = cv2.imread('DATA/7_sultan_ahmet.jpg',0)
ret1,th1 = cv2.threshold(resim,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

plt.imshow(th1,cmap="gray")
cv2.imwrite("deneme.jpg",th1)
```

### Blur

```python
resim1 = cv2.imread('DATA/6_noised_pepber.jpg')

resim1 = cv2.cvtColor(resim1, cv2.COLOR_BGR2RGB)

sonuc1 = cv2.blur(resim1,ksize=(3,3))
sonuc2 = cv2.medianBlur(resim1,ksize=3)
sonuc3 = cv2.GaussianBlur(resim1,ksize=(3,3),sigmaX=50)
sonuc4 = cv2.bilateralFilter(resim1,d=3,sigmaColor=20,sigmaSpace=20)
sonuc5 = cv2.boxFilter(resim1,ddepth=-1,ksize=(3,3))
```

```python
pencere = plt.figure(figsize=(15,10))
pencere.add_subplot(2,3,1)
plt.imshow(resim1,)# cmap='gray')
pencere.add_subplot(2,3,2)
plt.imshow(sonuc1,)# cmap='gray')
pencere.add_subplot(2,3,3)
plt.imshow(sonuc2,)# cmap='gray')
pencere.add_subplot(2,3,4)
plt.imshow(sonuc3,)# cmap='gray')
pencere.add_subplot(2,3,5)
plt.imshow(sonuc4,)# cmap='gray')
pencere.add_subplot(2,3,6)
plt.imshow(sonuc5,)# cmap='gray')
```
